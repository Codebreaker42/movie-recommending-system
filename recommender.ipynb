```python

def convert(text):
    list=[]
    for txt in ast.literal_eval(text):
        list.append(txt['name'])
    return list
```

```python
def convert_cast(text):
    list=[]
    counter=0
    for actors in ast.literal_eval(text):
        if counter !=5:
            list.append(actors['name'])
            counter=counter+1
        else:
            break
    return list        
        
```

```python
def fetch_director(text):
    list=[]
    for director in ast.literal_eval(text):
        if(director['job']=='Director'):
            list.append(director['name'])
            break
    return list
```

# removing spaces to solve name conflict

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(str).replace(" ","")
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

# make new columns tag

```python
movies.info()
# movies['tags']=movies['overview']+movies['genres']+movies['cast']+movies['crew']+movies['original_language']+movies['release_year']
# movies['tags'] += movies['keywords'].apply(lambda x: ' '.join(x))

# Define the columns you want to include in the 'tags' column
columns_to_include = ['overview', 'genres', 'keywords', 'cast', 'crew', 'original_language', 'release_year']

# Convert list elements to strings and concatenate the values from the selected columns
movies['tags'] = movies[columns_to_include].apply(lambda x: ' '.join(map(str, x)), axis=1)
# Optionally, you can remove any extra spaces
movies['tags'] = movies['tags'].str.strip()

def clean_tags(tags_str):
    # Remove square brackets and single quotes
    tags_cleaned = re.sub(r"[\[\]']", '', tags_str)
    # Split the cleaned string into a list of tags
    tags_list = tags_cleaned.split(', ')
    # Join the list elements into a single string with spaces
    tags_string = ' '.join(tags_list)
    return tags_string

# Apply the clean_tags function to the 'tags' column
movies['tags'] = movies['tags'].apply(clean_tags)
```

# NLP

```python
movies=movies[['movie_id','title','tags']]
movies.head()
```

```python
# 1-lower case
movies['tags']=movies['tags'].str.lower()
movies['tags'][1]
```

```python
# removing punctuation
import string 
punc=string.punctuation
print(punc)
def remove_punc(text):
    for char in punc:
        text=text.replace(char,'')
    return text
movies['tags']=movies['tags'].apply(remove_punc)
movies['tags'][1]
```

```python
# removing stopwords
from sklearn.feature_extraction import text

def remove_stopwords(txt):
    # Split the text into words
    words = txt.split()
    # Remove stop words and join the remaining words with spaces
    filtered_words = [word for word in words if word not in text.ENGLISH_STOP_WORDS]
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text

# remove_stopwords("a long time here is a man name nitin ")
movies['tags']=movies['tags'].apply(remove_stopwords)
movies['tags'][1]
```

```python
# stemming 
# !pip install nltk
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
# stem_words("dancing playing ")
movies['tags'] = movies['tags'].apply(stem_words)
```

# text vectorization(bow)

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
vectors=bow.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
# bow.get_feature_names()
```

# model training

```python
# cosine distance
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend_movies(movie_name):
    movie_index=movies[movies['title']==movie_name].index[0]
    distances=similarity[movie_index]
    recommend_movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for recommend_movies in recommend_movie_list:
#         print(recommend_movies[0])
        print(recommend_movies[0])
#         print(movies.iloc[recommend_movies[0]].title)
```
